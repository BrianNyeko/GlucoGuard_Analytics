Before taking any decision about the best strategy to handle with the data, we shoudl understand the nature of those attributes: 

= **max_glu_serum** feature refers to the maximum glucose serum test result recorded for a patient. Glucose serum tests measure the concentration of glucose (sugar) in the blood, and they are commonly used to diagnose and monitor diabetes.

Some of the values that appear, '300' or '200' for example, give us a lack of visibility for that variable. So imputation is not a good strategy becuase it could include a big bias in the dataframe. For that reason, all the "none" results will be understood as the test was not performed/not measured.  

= **A1Cresult** feature  refers to the result of the A1C test, also known as the hemoglobin A1c test. This test measures the average blood sugar level over the past 2 to 3 months and is commonly used to diagnose and monitor diabetes. The result of this test could be Normal/Negative (suggesting good blood sugar control), if no abnormality is detected, then the result is 'Norm'. The test result can be Abnormal/Positive (suggesting poor blood sugar control), in the dataset, this might be represented by specific values indicating different levels of abnormality, such as ">7%" or ">8%", which typically represent elevated A1C levels associated with diabetes. Or simply the test result can be Not measured. 

So, here what I suggest would be transforming all the values in three categories: 

   - Category 1: Positive = "norm"
   
   - Category 2: Negative = ">7%" or ">8%"
  
   - Category 3: Not measured = "none"

```python
# Dealing with missing values

# Define a function to transform the values
unique_values1 = df['A1Cresult'].unique()
print("Unique values of the 'A1Cresult' attribute:", unique_values1)

def transform_A1Cresult(value):
    if value == 'Norm':
        return 'Negative'
    elif value in ['>7', '>8']:
        return 'Positive'
    else:
        return 'Not measured'

unique_values2 = df['max_glu_serum'].unique()
print("Unique values of the 'max_glu_serum:", unique_values2)

def transform_max_glu_serum(value):
    if value == 'Norm':
        return 'Negative'
    elif value == '>300':
        return 'High hyperglycemia'
    elif value == '>200':
        return 'Hyperglycemia'
    else:
        return 'Not measured'

# For A1C result: 

# Apply the transformation function to the 'A1Cresult' column
df['A1Cresult_transformed'] = df['A1Cresult'].apply(transform_A1Cresult)
category_counts = df['A1Cresult_transformed'].value_counts()

# Display the counts
print("Positive:", category_counts['Positive'])
print("Negative:", category_counts['Negative'])
print("Not measured:", category_counts['Not measured'])

# Display the transformed column
# print(df[['A1Cresult', 'A1Cresult_transformed']])

# For max_glu_serum: 

    # Apply the transformation function to the 'max_glu_serum' column
df['max_glu_serum_trasnformed'] = df['max_glu_serum'].apply(transform_max_glu_serum)

cc = df['max_glu_serum_trasnformed'].value_counts()

# Display the counts
print("High hyperglicemia - Urgent medical attention:", cc['High hyperglycemia'])
print("Hyperglicemia:", cc['Hyperglycemia'])
print("Not measured:", cc['Not measured'])
print("Normal values:", cc['Negative'])

# Display the transformed column
# print(df[['max_glu_serum', 'max_glu_serum_trasnformed']])

#FEATURE SELECTION

# List of features you want to select
selected_features = ['gender','race','admission_type_id','readmitted','time_in_hospital', 'metformin','insulin','num_lab_procedures', 'num_procedures', 'num_medications', 'number_diagnoses', 'number_emergency']

# Select the list of features from the DataFrame
selected_df = df[selected_features]

# Display the selected DataFrame
# print(selected_df)
```
